```python
# Function to Remove URLs
def remove_url(txt):
    #Replace URLs found in a text string with nothing 
    """ 
    Parameters ----> txt : string
    Returns -------> The txt string with url's removed.
    """
    return " ".join(re.sub("([^0-9A-Za-z \t])|(\w+:\/\/\S+)", "", txt).split())
```

```python
tweets_minus_urls = [remove_url(tweet) for tweet in data['Content']]
```

```python
data['Tweets w/o URLs'] = pd.Series(tweets_minus_urls)
```

```python
custom_stop_word_list = ['a','we','the','of','and','has','if','is','to','my','your','should','cant','wouldn\'t','his','her','they', 'so','what','that','while']
```

```python
search_array = data['Search Term'].unique()
```

```python
cloud_data = []
for i in range(len(search_array)):
    cloud_data.append(data['Tweets w/o URLs'][data['Search Term']==search_array[i]])
```

```python
# Tweets From the First Search Term
cloud_data[0].head()
```

```python
# Function used to eliminate some excess filler words for the purpose of creating meaningful wordclouds
def extract_cloud_words(tweets_data, wordlist):
    '''
    Parameters: Series of Tweets, List of Stop Words
    
    Returns: List of all words in tweets minus the stop words contained in the tweets
    '''
    cloud_words = []
    for tweet in tweets_data:
        words = tweet.lower().split()
        for word in words:
            if word not in wordlist:
                cloud_words.append(word)
    return cloud_words    
```

```python
# Creating and plotting our wordclouds
for i in range(len(cloud_data)):
    corpus = extract_cloud_words(cloud_data[i], custom_stop_word_list)
    text = ' '.join(corpus)

    # Create and generate the 22 Word Clouds
    wordcloud = WordCloud().generate(text)
    
    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis("off")
    plt.show()
```

```python
# Sentiment Analysis Function
def analyze_sentiment(tweet):
    '''
    function used to classify the polarity of a tweet using textblob.
    '''
    analysis = TextBlob(tweet)
    if analysis.sentiment.polarity > 0:
        return 1
    elif analysis.sentiment.polarity == 0:
        return 0
    else:
        return -1
    
# Future change note: consider just returning analysis.sentiment.polarity instead
```

```python
# Tweets w/o URLs
data['SA'] = np.array([analize_sentiment(tweet) for tweet in data['Tweets w/o URLs']])
data.head(15)
```
